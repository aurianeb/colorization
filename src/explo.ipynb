# Imports

```python
import matplotlib.pyplot as plt
import SimpleITK as sitk
%matplotlib inline
```

# Loading

```python
img = sitk.ReadImage('data/BRATS2015_Training/HGG/brats_2013_pat0001_1/VSD.Brain.XX.O.MR_Flair.54512/VSD.Brain.XX.O.MR_Flair.54512.mha')
```

# Plotting

```python
arr_img = sitk.GetArrayViewFromImage(img)
arr_img.shape
```

```python
fig, ax = plt.subplots(3, 5, figsize=(12, 6))
for i, ax in enumerate(fig.axes):
    _ = ax.set_title('z = {}'.format(i * 10))
    _ = ax.imshow(arr_img[i * 10, :, :])

_ = plt.tight_layout()
```

```python
def myshow(img):
    nda = sitk.GetArrayViewFromImage(img)
    plt.imshow(nda)
```

```python
size = img.GetSize()
size
myshow(img[:, :, size[2] // 2])
```

```python
nslices = 10
slices = [img[:, :, s] for s in range(0, size[2], size[0] // (nslices + 1))]

_ = plt.figure(figsize=(12, 9))
myshow(sitk.Tile(slices, [1, 0]))
```
